```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
plt.plot(x,y,'r')
plt.xlabel("x Achse")
plt.ylabel("y Achse")
plt.title("Diagramm Titel")#
#plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.plot(y,x, 'g*-')
```

```python
af = plt.figure()

axes = af.add_axes([0.1,0.1,0.5,0.8])

axes.plot(x,y,'b')

axes.set_xlabel('X Label')

axes.set_ylabel('Y Label')

axes.set_title('Überschrift')
```

```python
af = plt.figure()

axes1 = af.add_axes([0.1,0.1,0.8,0.8])

axes2 = af.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes.set_title('GROSSE')

axes2.plot(y,x,'r')
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("KLEIN")
```

```python
diag, axes = plt.subplots(nrows = 1, ncols = 2)

for i in axes:
    print(i)
    i.plot(x,y)
    i.set_xlabel('x')
    i.plot(x,y,'r')
    i.set_xlabel('X')
    i.set_ylabel('Y')
    i.set_title("Titel")
    
plt.tight_layout()    
```
